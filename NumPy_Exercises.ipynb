```python
a = np.random.randint(10, size = (5,5))
a
```

```python
alt_a[0,0] = 1000
alt_a[1,1] = 1000
alt_a
```

```python
a
```

```python
# Subset effects main set
```

```python
alt_b = a[0:3,0:2].copy()
alt_b
```

```python
alt_b[0,0] = 0
alt_b[1,1] = 0
alt_b
```

```python
a
```
